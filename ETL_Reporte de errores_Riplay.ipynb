```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
from zoneinfo import ZoneInfo
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
pd.set_option('display.max_columns', None)
```

```python
df_desgravamen= pd.read_excel('C:/DATOS/Ripley Todo Desgravamen.xlsx', sheet_name=0,
                              dtype={'NRO LOTES ANTERIORES': str, 'NRO. LOTE DEL ERROR': str, 
                                     'CÓDIGO ERROR': str, 'PLACA': str, 
                                     'MOTOR': str, 'TIPO DOC. ID': str, 'DOC. ID.': str})
df_desgravamen.columns = (df_desgravamen.columns.str.strip().str.upper()
                            .str.replace(r'[^A-Za-z0-9]', '_', regex=True))

#df_desgravamen.drop(['SUMA'], axis=1, inplace=True)
```

```python
df_desgravamen= df_desgravamen.rename(columns={'C_DIGO_PRODUCTO':'CODIGO_PRODUCTO','C_D__CERTIFICADO_CANAL':'CODIGO_CERTIFICADO',
                                               'C_DIGO_ERROR':'CODIGO_ERROR', 'DESCRIPCI_N_ERROR':'DESCRIPCION_ERROR',
                                                'VALIDACI_N_APLICADA':'VALIDACION_APLICADA', 'VALIDACI_N_APLICADA':'VALIDACION_APLICADA',
                                                'NRO__LOTE_DEL_ERROR':'NRO_LOTE_DEL_ERROR', 'DOC__ID_':'DOC_ID', 'TIPO_DOC__ID': 'TIPO_DOC_ID',
                                                'FEC__INICIO':'FECHA_INICIO', 'FEC__FIN':'FECHA_FIN', 'L_NEA_DE_TRAMA':'LINEA_DE_TRAMA',
                                                'PRIMABRUTACAN':'PRIMA_BRUTA', 'PRIMANETACAN':'PRIMA_NETA', 
                                                'LINEA_TRAMA_TEXTO_COMPLETO_':'TRAMA_TEXTO_COMPLETO'})
```

```python
df_desgravamen.head(3)
```

```python
df_desgravamen['PRODUCTO'] = df_desgravamen["PRODUCTO"].str.upper()
df_desgravamen['TIPO_MOVIMIENTO'] = df_desgravamen["TIPO_MOVIMIENTO"].str.upper()
df_desgravamen['ORIGEN_ERROR'] = df_desgravamen["ORIGEN_ERROR"].str.upper()
df_desgravamen['CODIGO_CERTIFICADO'] = df_desgravamen["CODIGO_CERTIFICADO"].str[1:]

df_desgravamen['FECHA_CARGA_LOTE']= pd.to_datetime(df_desgravamen['FECHA_CARGA_LOTE'], errors='coerce').dt.date
df_desgravamen['FECHA_NACIMIENTO']= pd.to_datetime(df_desgravamen['FECHA_NACIMIENTO'], format='%Y-%m-%d', errors='coerce').dt.date
df_desgravamen['FECHA_INICIO']= pd.to_datetime(df_desgravamen['FECHA_INICIO'], errors='coerce').dt.date
df_desgravamen['FECHA_FIN']= pd.to_datetime(df_desgravamen['FECHA_FIN'], errors='coerce').dt.date
```

```python
df_desgravamen["TIPO_SEGURO"] = df_desgravamen["TRAMA_TEXTO_COMPLETO"].str[:3]
df_desgravamen["CODIGO_MOVIMIENTO"] = df_desgravamen["TRAMA_TEXTO_COMPLETO"].str[48]
df_desgravamen["TIPO_REGISTRO"] = df_desgravamen["TRAMA_TEXTO_COMPLETO"].str[43:45]

df_desgravamen['FECHA_CARGA'] = datetime.now(ZoneInfo("America/Lima")).strftime("%Y-%m-%d")
df_desgravamen['FECHA_CARGA'] = pd.to_datetime(df_desgravamen['FECHA_CARGA'], errors='coerce').dt.date
```
